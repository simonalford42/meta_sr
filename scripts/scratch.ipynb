```python
from evaluation import analyze_r2_across_runs
```

```python
all_scores = analyze_r2_across_runs()
```

```python
all_scores
```

```python
splits = ['splits/srbench_all.txt', 'splits/split_train.txt', 'splits/split_val.txt']

for split in splits:
    with open(split, 'r') as f:
        tasks = [line.strip() for line in f.readlines()]

    split_scores = {t: all_scores[t] for t in tasks}

    scores = []
    for i in range(5):
        task_scores_dict = {t: split_scores[t][i] for t in split_scores}
        scores.append(task_scores_dict)
    avgs = []
    for i in range(5):
        avg = sum(scores[i].values()) / len(scores[i])
        avgs.append(avg)

    avg_avg = sum(avgs) / len(avgs)
    std_avg = (sum((x - avg_avg) ** 2 for x in avgs) / len(avgs)) ** 0.5
    avgs_str = ','.join([f"{x:.2f}" for x in avgs])
    print(f"mean/std for avg R² across runs in split {split}: {avg_avg:.4f} ± {std_avg:.4f} ({avgs_str})")
```
